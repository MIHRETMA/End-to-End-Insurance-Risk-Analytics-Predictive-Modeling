Import Necessary Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/MachineLearningRating_v3.txt', sep='|') # Load data
df.head() # Display first few rows
```

```python
df.columns # List column names
```

```python
df["TransactionMonth"]= pd.to_datetime(df["TransactionMonth"], errors='coerce') # Convert to datetime
```

Calculating Loss Ratio and Grouping them by Important Variables

```python
df["LossRatio"] = df["TotalClaims"] / df["TotalPremium"] # Calculate Loss Ratio
df.groupby("Province")["LossRatio"].mean().sort_values(ascending=False) # Loss Ratio by Province
df.groupby("VehicleType")["LossRatio"].mean().sort_values(ascending=False) # Loss Ratio by Vehicle Type
df.groupby("Gender")["LossRatio"].mean().sort_values(ascending=False) # Loss Ratio by Gender
```

Creating Plots

```python
print(df["CustomValueEstimate"].isna().sum()) # Check missing values in CustomValueEstimate
```

```python
plt.hist(df["TotalPremium"], bins=50) # Premium Distribution 
plt.title("Premium Distribution")
plt.savefig("../plots/premium_dist.png") # Save the histogram
plt.close()
plt.hist(df["TotalClaims"], bins=50) # Claim Distribution 
plt.title("Claim Distribution")
plt.savefig("../plots/claim_dist.png") # Save the histogram
plt.close()
plt.hist(df["CustomValueEstimate"], bins=50) # CustomValueEstimate Distribution 
plt.title("CustomValueEstimate Distribution")
plt.savefig("../plots/CustomValueEstimate_dist.png") # Save the histogram
plt.close()
plt.boxplot(df["TotalClaims"]) # Claims Outlier Check
plt.title("Claims Outlier Check")
plt.savefig("../plots/claims_outliers.png") # Save the boxplot
plt.close()
plt.boxplot(df["CustomValueEstimate"]) # CustomValueEstimate Outlier Check
plt.title("CustomValueEstimate Outlier Check")
plt.savefig("../plots/CustomValueEstimate_outliers.png") # Save the boxplot
plt.close()
```

```python
monthly = df.groupby(df["TransactionMonth"].dt.to_period("M")).agg({
   "TotalClaims": "sum",
   "TotalPremium": "sum"
}) # Monthly aggregation
monthly.plot() # Plot monthly trends
plt.title("Monthly Claim & Premium Trend")
plt.savefig("../plots/monthly_trend.png") # Save the plot
plt.close()
```

Top LossRatio per make

```python
df["LossRatio"] = df["LossRatio"].replace([np.inf, -np.inf], 0) # Replace infinite values with 0
df.groupby("make")["LossRatio"].sum().sort_values(ascending=False).head(10) # Top 10 Makes by Loss Ratio
```

Bottom LossRatio per make

```python
df.groupby("make")["LossRatio"].sum().sort_values().head(10) # Bottom 10 Makes by Loss Ratio
```
